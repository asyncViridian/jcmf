```python
import matplotlib.pyplot as plt
import random
import math

with open('../output/scorelistings/modified-null.txt') as f:
    nulls = f.read().splitlines()
with open('../output/scorelistings/modified-real.txt') as f:
    reals = f.read().splitlines()
nullzip = [(float(i), 0) for i in nulls]
realzip = [(float(i), 1) for i in reals]
alldata = nullzip + realzip
random.shuffle(alldata)
ordered = sorted(alldata, key=lambda tup: tup[0])

# the thresholds used
x = []
# the false positive rate
y = []
truereal = len(reals) + 1
falsereal = len(nulls) + 1
truenull = 1
falsenull = 1
lastval = 0
for tup in ordered:
    if (tup[0] != lastval):
        # write a new point
        x = x + [lastval]
        y = y + [(falsereal / (falsereal + truenull))]
        print(str(lastval) +"\t" + str(falsereal / (falsereal + truenull)))
        #y = y + [math.log(falsereal / (falsereal + truenull))]
        lastval = tup[0]
    if (tup[1] == 1):
        # this is now predicted to be false but is actually true
        truereal -= 1
        falsenull += 1
    else:
        # this is now predicted to be false and is actually false
        falsereal -= 1
        truenull += 1

plt.plot(x, y)
```
